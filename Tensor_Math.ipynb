```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
#Addition
tensor_a + tensor_b
```

```python
#Subtraction
tensor_b - tensor_a
```

```python
#Multiplication
tensor_a * tensor_b
```

```python
#Division
tensor_b / tensor_a
```

```python
#Remainders
tensor_b % tensor_a
```

```python
#Exponents
torch.pow(tensor_a, tensor_b)
```
